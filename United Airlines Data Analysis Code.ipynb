```python
import pandas as pd

# Load dataset
df = pd.read_csv('calls.csv')

# Calculate Handling Time and making a new column of it
df['HandleTime'] = pd.to_datetime(df['call_end_datetime']) - pd.to_datetime(df['agent_assigned_datetime'])
df['HandleTime']
```

```python
#checking that new column is loaded
df
```

```python
# Merging the dataset
df2 = pd.read_csv('sentiment_statistics.csv')
merged_df = pd.merge(df, df2, on='call_id')
```

```python
merged_df
```

```python
# Calculating Waiting Time and making a new Column of it
merged_df['WaitingTime'] = pd.to_datetime(df['agent_assigned_datetime']) - pd.to_datetime(df['call_start_datetime'])
merged_df['WaitingTime']
```

```python
# Dropped some columns that are no longer needed
df_cleaned = merged_df.drop(['call_start_datetime', 'agent_assigned_datetime', 'call_end_datetime', 'agent_id_y'], axis=1)
df_cleaned
```

```python
# Merging the dataset
df_reason = pd.read_csv('reason.csv')
df_final = pd.merge(df_cleaned, df_reason, on='call_id')
```

```python
df_final
```

```python
# Calculating average handling time wrt the primary call reason
aht_callreason = df_final.groupby('primary_call_reason')['HandleTime'].mean()
```

```python
aht_callreason
```

```python
# Cleaning the primary call reason column to remove the white spaces and other unneccesary characters
df_final['primary_call_reason']= df_final['primary_call_reason'].str.strip().str.lower()

# Want to remove all extra spaces between words:
df_final['primary_call_reason']= df_final['primary_call_reason'].str.replace('\s+', ' ', regex=True).str.lower()
```

```python
df_final['primary_call_reason'].unique()
```

```python
# Cleaning the primary call reason column to remove other unneccesary characters
df_final['primary_call_reason']= df_final['primary_call_reason'].str.replace('-', ' ', regex=True).str.lower()
df_final['primary_call_reason']= df_final['primary_call_reason'].str.replace('&', 'and', regex=True).str.lower()
```

```python
# Verifying that the column is clean
df_final['primary_call_reason'].unique()
```

```python
# Calculating average handling time wrt the cleaned primary call reason column
aht_callreason = df_final.groupby('primary_call_reason')['HandleTime'].mean()
```

```python
aht_callreason
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Making a graph to look for the pattern in AHT and Primary Call Reason to look for improvement in specific domains
df_final['HandleTime_seconds'] = df_final['HandleTime'].dt.total_seconds()

# Plot AHT by PrimaryCallReason
plt.figure(figsize=(10, 4))  # Adjust figure size for readability
sns.barplot(x='primary_call_reason', y='HandleTime_seconds', data=df_final)
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.title('Average Handle Time (in seconds) by Primary Call Reason')
plt.xlabel('Primary Call Reason')
plt.ylabel('Average Handle Time (seconds)')
plt.show()
```

```python
#Calculate the frequency of each CallReason
call_reason_frequency = df_final['primary_call_reason'].value_counts()

# Step 3: Identify the most and least frequent CallReasons
most_frequent_call_reason = call_reason_frequency.idxmax()  # Most frequent
least_frequent_call_reason = call_reason_frequency.idxmin()  # Least frequent
```

```python
most_frequent_call_reason
```

```python
least_frequent_call_reason 
```

```python
df_final
```

```python
# Filter the dataframe for most and least frequent CallReasons
most_frequent_df = df_final[df_final['primary_call_reason'] == most_frequent_call_reason]
least_frequent_df = df_final[df_final['primary_call_reason'] == least_frequent_call_reason]


# Calculate AHT for most and least frequent CallReasons
aht_most_frequent = most_frequent_df['HandleTime_seconds'].mean()
aht_least_frequent = least_frequent_df['HandleTime_seconds'].mean()

# Output the results
print(f"Most Frequent Call Reason: {most_frequent_call_reason} with AHT: {aht_most_frequent} seconds")
print(f"Least Frequent Call Reason: {least_frequent_call_reason} with AHT: {aht_least_frequent} seconds")
```

```python
# Step 1: Calculate the absolute difference between the two AHT values
aht_difference = abs(aht_most_frequent - aht_least_frequent)

# Step 2: Calculate the percentage difference
percentage_difference = (aht_difference / aht_most_frequent) * 100

# Output the result
print(f"Percentage Difference between AHT of most and least frequent call reasons: {percentage_difference:.2f}%")
```

```python
# Check the influence of average sentiment of the call on AHT
# Create bins for sentiment score in intervals of 0.25
df_final['sentiment_bin'] = pd.cut(df_final['average_sentiment'], bins=pd.interval_range(start=-1, end=1, freq=0.25))

# Calculate average AHT for each sentiment bin
avg_aht_by_sentiment = df_final.groupby('sentiment_bin', observed=True)['HandleTime_seconds'].mean()

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# Visualize the comparison between sentiment bins and AHT
plt.figure(figsize=(10, 4))
sns.barplot(x='sentiment_bin', y='HandleTime_seconds', data=df_final, palette='coolwarm')
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.title('Average Handle Time (AHT) by Sentiment Score Range')
plt.xlabel('Sentiment Score Range')
plt.ylabel('Average Handle Time (seconds)')
plt.show()
```

```python
# Count occurrences of each primary_call_reason
call_reason_counts = df_final['primary_call_reason'].value_counts()

# Create the new column by mapping the counts to the original DataFrame
df_final['call_reason_frequency'] = df_final['primary_call_reason'].map(call_reason_counts)
```
